### Correlation
    There are some very strong correlations as well and some seemingly overpowering features when it comes to how they affect the country's overall happiness.

##### Positive correlation
    Features with a strong positive correlation to eachother are: 'Life Ladder', 'Log GDP per capita', 'Social support', and 'Healthy life expectancy at birth'. Two others with slightly weaker positive correlation are 'Positive affect' and 'Freedom to make life choices'.
    
    This means that when one of these features is higher, the others are likely too, which shows that happy countries generally have higher numbers in all of these.

##### Negative correlation
    Features with a weak to medium negative correlation to everything but eachother are: 'Perceptions of corruption' and 'Negative affect'.

    This means that when these have higher values, the other features tend to have lower values, meaning when these are high the general happiness is usually lower.

##### No correlation
    The feature 'Generosity' surprisingly doesn't have much of a correlation with anything.

    This means that the happiness of a country doesn't really affect their generosity.
### Improvements
    To improve the dataset, it could be beneficial to include an objective measure of well-being, data from tracking the same individuals over time, more features such as mental health, and perhaps scores tracked by smaller regions instead of just country.

```python
# Reset df between runs to not create too many columns
df = pd.read_excel('DataForTable2.1WHR2023.xls')

# Function to normalize data so features with larger scales don't outweigh the others
def min_max_normalize(column):
    min_value = np.min(column)
    max_value = np.max(column)
    normalized = (column - min_value) / (max_value - min_value)
    return normalized

# Apply normalization
for column in df.columns:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64':
        if column not in ['Country name', 'year']:
            df[column + '_normalized'] = min_max_normalize(df[column])

# Function to calculate the happiness score for a row based on the normalized features, subtracting the negatively correlated
def calculate_happiness_score(row):
    positive_features = ['Life Ladder_normalized', 'Log GDP per capita_normalized', 'Social support_normalized', 
                'Healthy life expectancy at birth_normalized', 'Freedom to make life choices_normalized', 
                'Generosity_normalized', 'Positive affect_normalized']
    positive_score = row[positive_features].sum()
    
    negative_features = ['Perceptions of corruption_normalized', 'Negative affect_normalized']
    negative_score = row[negative_features].sum()
    
    score = positive_score - negative_score
    
    return score / len(positive_features)

# Calculate happiness for each row
df['Overall Happiness Score'] = df.apply(calculate_happiness_score, axis=1)

df.describe()
```

## 3) Most Changed Happiness
    This compares the first year of data collected for the country, to their 2022 data, which inherently excludes all countries without 2022 data.

```python
# Filter countries without data for 2022
df_2022 = df[df['year'] == 2022]
countries_with_2022_data = df_2022['Country name'].unique()
df_filtered = df[df['Country name'].isin(countries_with_2022_data)]

# Find first year data for all countries
first_year_data = df_filtered.groupby('Country name').first().reset_index()
merged_df = pd.merge(first_year_data[['Country name', 'Overall Happiness Score']], 
                     df_2022[['Country name', 'Overall Happiness Score']], 
                     on='Country name', suffixes=('_first', '_2022'))

# Calculate change in overall happiness score
merged_df['Change in Overall Happiness'] = merged_df['Overall Happiness Score_2022'] - merged_df['Overall Happiness Score_first']

merged_df.describe()
```

### Most Increased

```python
merged_df[merged_df['Change in Overall Happiness'] == merged_df['Change in Overall Happiness'].max()]
```

```python
merged_df[merged_df['Change in Overall Happiness'] == merged_df['Change in Overall Happiness'].min()]
```
